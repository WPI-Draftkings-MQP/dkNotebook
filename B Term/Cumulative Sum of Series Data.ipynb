```python
import pandas as pd
import numpy as np
import os
from ipywidgets import IntProgress
from IPython.display import display
```

```python
df = pd.read_csv('data/CombinedSeries.csv')
df = df.set_index('ContestId')
backup = df
backup.head()
```

```python
contestsBar = IntProgress(min=0, max=len(df), description='Contests:')
display(contestsBar)

for index, contest in df.iterrows():
    cumsum = 0
    for column in reversed(df.columns):
        cumsum += contest[column]
        contest[column] = cumsum
    contestsBar.value += 1
```

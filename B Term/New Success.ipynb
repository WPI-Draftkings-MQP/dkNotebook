```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
contests = pd.read_csv('data/WorkingData.csv').set_index('ContestId')
```

```python
# contests[['MaxNumberPlayers', 'MaxEntriesPerUser', 'EntryFeeAmount', 'Entries']]
contests = contests.assign(Success = (contests['Entries'] == contests['MaxNumberPlayers']))
len(contests.loc[contests['Success']==True])/len(contests)
```

```python
contests = contests.assign(FillGapDollar = (contests['MaxNumberPlayers'] - contests['Entries'])*contests['EntryFeeAmount'])
```

```python
contests = contests[['Success', 'FillGapDollar']]
```

```python
contests.to_csv('data/SuccessAndFillGapDollars.csv')
```

```python
fullContests = pd.read_csv('data/WorkedData.csv').drop(columns=['Success'], axis=2).set_index('ContestId')
```

```python
fullContests.head()
```

```python
combinedContests = pd.merge(contests, fullContests, on='ContestId', how='left')
```

```python
combinedContests.head()
```

```python
columns = []
for col in combinedContests.columns:
    if "Payout_" not in col:
        columns.append(col)
combinedContests = combinedContests[columns]
```

```python
combinedContests.to_csv('data/WorkedData.csv')
```

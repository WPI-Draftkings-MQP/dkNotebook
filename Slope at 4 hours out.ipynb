```python
chunkMapDF = pd.read_csv('data/ChunkMap.csv')
```

```python
slopes = []
for chunkName in chunkMapDF['Chunk'].unique()[:]:
    print(chunkName)
    chunk = chunkMapDF[chunkMapDF['Chunk'] == chunkName]
    chunkDF = pd.read_csv('data/chunks/'+chunkName+".csv").drop(columns=['Unnamed: 0'])
    for cid in chunk['ContestId'][:]:
                
        numMeasures = 5

        
        df = chunkDF[chunkDF['ContestId']==cid]
        df = df.assign(SecondsTill=df['SecondsRemaining']-1440)
        df = df[df['SecondsTill'] >= 0].sort_values(by='SecondsTill').reset_index()
        if(len(df) < numMeasures):
            continue;
        else:
            final = df.iloc[0]
            entrySum = final.Entries
            timePassed = final.SecondsTill

        if(len(df) < numMeasures):
            numMeasures = len(df)
    #         if(len(df) >=numMeasures):
    #             print(df.head(numMeasures))
        for i in range(1, numMeasures):
            curr = df.iloc[i]
            entrySum+= curr['Entries']
            timePassed = curr['SecondsTill']
        if(timePassed == 0):
            timePassed = 1
        slope = entrySum/timePassed
        slopes.append({'ContestId':final.ContestId, 'InstantaneousSlope':final.Entries, 'AvgSlope':slope})
```

```python
slopeDF = pd.DataFrame(data=slopes)
```

```python
successes = pd.read_csv('data/WorkedData.csv').set_index('ContestId')[['Success']]
```

```python
pd.merge(slopeDF, successes, on='ContestId', how='left').set_index('ContestId').dropna().to_csv('SlopesForJackson.csv')
```

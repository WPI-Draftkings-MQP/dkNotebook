```python
import numpy as np
import pandas as pd
```

```python
originalData = pd.read_csv('originalContests.csv')
```

```python
originalData.head()
```

```python
originalData['is_vault'] = np.random.uniform(0, 1, len(originalData)) <= .2
vault, workingData = originalData[originalData['is_vault']==True], originalData[originalData['is_vault']==False]
print('Number of observations in the vault data:', len(vault))
print('Number of observations in the working data:',len(workingData))
```

```python
vault.to_csv('VaultData.csv')
workingData.to_csv('WorkingData.csv')
```
